```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
```

```python
ds = pd.read_csv("president_heights.csv") # dataset
print(ds.head()) # print first 5 rows
```

```python
height = np.array(ds["height(cm)"])
print(height)
```

```python
print(ds.info())
```

```python
# look at some numbers
# https://www.w3schools.com/python/python_ml_mean_median_mode.asp

print("Mean of heights:", np.mean(height)) # the average value
print("Mode:", stats.mode(height)) # the most common value
print("Standart deviation:", np.std(height)) # number that describes how spread out the values are
print("Variance:", np.var(height))
print("Minimum height:", height.min())
print("Maximum height:", height.max())
```

```python
print("25th percentile:", np.percentile(height, 25))
print("Median:", np.median(height)) # we see that the median is 182.0 cm
print("75th percentile:", np.percentile(height, 75))
```

```python
sns.set()
```

```python
plt.hist(height)
plt.title("Height Distribution of US Presidents")
plt.xlabel("height(cm)")
plt.ylabel("number")
plt.show()
```
